```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create necessary folders
folders = [
    "project/data/student_performance/raw",
    "project/data/student_performance/processed",
    "project/figures/student_performance",
    "project/src",
    "project/results",
    "project/reports"
]
for folder in folders:
    os.makedirs(folder, exist_ok=True)

# Load the student performance dataset
data = pd.read_csv("StudentsPerformance.csv")  # Ensure correct filename
data.to_csv("project/data/student_performance/raw/student_performance.csv", index=False)
```

## 1. Histogram - Math Score Distribution  
This histogram helps visualize the distribution of math scores among students.  
- It shows how frequently different score ranges occur.  
- The KDE (Kernel Density Estimate) curve helps observe the overall trend.  
- This allows us to identify whether scores are normally distributed or skewed.  


```python
# 1. Histogram - Math Score Distribution
plt.figure(figsize=(8,5))
sns.histplot(data['math score'], bins=10, kde=True)  # Correct
plt.title("Math Score Distribution")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.savefig("project/figures/student_performance/histogram_math_scores.png")
plt.tight_layout()
plt.show()
```

## 2. Boxplot - Gender vs. Math Score  
A boxplot provides insights into the variation of math scores between male and female students.  
- It highlights **median scores, interquartile ranges (IQR), and outliers.**  
- We can compare performance differences across genders.  
- Outliers help identify students with exceptionally high or low scores.  


```python
# 2. Boxplot - Gender vs. Math Score
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='math score', data=data)  # Correct
plt.title("Math Score Variation by Gender")
plt.savefig("project/figures/student_performance/boxplot_gender_math.png")
plt.tight_layout()
plt.show()
```

## 3. Scatter Plot - Reading Score vs. Writing Score  
This scatter plot shows the relationship between reading and writing scores.  
- A **positive correlation** means students who score high in reading also perform well in writing.  
- If points form an upward trend, it confirms a **strong correlation**.  
- This helps us understand how related these two skills are.  


```python
# 3. Scatter Plot - Reading Score vs. Writing Score
plt.figure(figsize=(8,5))
sns.scatterplot(x='reading score', y='writing score', data=data)  # Correct column names
plt.title("Reading Score vs. Writing Score")
plt.savefig("project/figures/student_performance/scatter_reading_vs_writing.png")
plt.tight_layout()
plt.show()
```

## 4. Bar Chart - Average Score per Subject  
A bar chart is used to compare the **average scores across Math, Reading, and Writing.**  
- It helps visualize which subject students perform best in on average.  
- If math scores are significantly lower, it may suggest areas needing improvement.  
- Educators can use this insight for curriculum adjustments.  


```python
# 4. Bar Chart - Average Score per Subject
plt.figure(figsize=(8,5))
data[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title("Average Score per Subject")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.savefig("project/figures/student_performance/bar_subject_scores.png")
plt.tight_layout()
plt.show()
```

## 5. Pie Chart - Test Preparation Course Distribution  
This pie chart illustrates the proportion of students who completed a test preparation course vs. those who did not.  
- It helps determine how many students took extra preparation.  
- We can further analyze if test prep correlates with higher scores.  
- This provides insight into the effectiveness of additional learning resources.  


```python
# 5. Pie Chart - Test Preparation Course Distribution
plt.figure(figsize=(8,5))
data['test preparation course'].value_counts().plot.pie(autopct="%1.1f%%", colors=['lightblue', 'lightgreen'])
plt.title("Test Preparation Course Distribution")
plt.ylabel("")
plt.savefig("project/figures/student_performance/pie_test_prep_distribution.png")
plt.tight_layout()
plt.show()
```
